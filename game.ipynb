```python
## the following line is just MAGIC ##############################################
## we need it to be able to draw images on the screen ############################
from medialib import *
initialize() # always the first instruction of the program
##################################################################################

print('OneDimensional game')
print('===================')
print('You are a lost hero, and have to find your way back home...')
print('reach the door to leave this level and get closer to home!\n')
print('To play: press a and d keys to move left and right')
print('         press s when the hero stand on pressure plate--This can switch on the bridge')
print('         press q to quit the game')

## game state
game_running = True
background='X-----P--wwwww-----E----X' ## the string representing the background
terrain = list(background) ## convert the string representing the background to a list
collidables = 'Xw' ## the letters that the avatar cannot pass
y_background = 200
player_pos = 3
y_player = 200
width_background = 32
wall = "OneDimensionalGame_imgs/X.png"
pressure_plate = "OneDimensionalGame_imgs/P.png"
pushed_pressure_plate = "OneDimensionalGame_imgs/_.png"
river = "OneDimensionalGame_imgs/w.png"
exit = "OneDimensionalGame_imgs/E.png"
floor = "OneDimensionalGame_imgs/-.png"
bridge = "OneDimensionalGame_imgs/=.png"
clear(245,245,245) ## background: almost white

while game_running:
    ## game initial state, draw the background
    ## the first image X is at (10,200)
    x_background = 10
    for i in terrain:
        if i=="X":
            draw(wall,x_background,y_background)
        elif i=="P":
            draw(pressure_plate,x_background,y_background)
        elif i=="w":
            draw(river,x_background,y_background)
        elif i=="E":
            draw(exit,x_background,y_background)
        elif i=="-":
            draw(floor,x_background,y_background)
        elif i=="=":
            draw(bridge,x_background,y_background)
        elif i=="_":
            draw(pushed_pressure_plate,x_background,y_background)
        x_background += width_background
    
    draw("OneDimensionalGame_imgs/H.png",10+player_pos*32,y_player)
    if terrain[player_pos] == "E":
        draw("OneDimensionalGame_imgs/you_win.png",100,50)
        wait(1)
        break
    userinput = wait_key_press()

    ## if the player push 'q', quit the game.
    if userinput == "q":
        break
        
    ## if the player push 'a',the avatar go backward（left）.
    ## update game state
    elif userinput == "a":
        if terrain[player_pos-1] not in collidables:
            player_pos -= 1
            
        
            
    ## if the player push 'd',the avatar go forward(right).
    ## update game state
    elif userinput == "d":
        if terrain[player_pos+1] not in collidables:
            player_pos += 1
        

    ## if the player push 's' while the avatar is at position P(pressure plate), then switch on the bridge.
    ## All images of water(w.png) will be changed to the image of bridges(=.png), and P.png will be changed to _.png.
    elif userinput == "s":
        if terrain[player_pos] == terrain[6]:
            terrain[9:14] = ["=","=","=","=","="]
            terrain[6] = "_"

    else:
        -1        

    
    clear(245,245,245)
clear(245,245,245)
text("Game Over",10,300,30,0,0,0)
wait_mouse_leftclick()
all_done()
```
